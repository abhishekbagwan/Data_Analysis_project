```python
# here we import the library 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read file

df_track=pd.read_csv('SpotifyFeatures.csv')
```

```python
 df_track.head()
```

```python
# cheak null values in a dataset

pd.isnull(df_track).sum()
```

```python
#To cheak all the columns

df_track.info()
```

```python
sorted_df=df_track.sort_values('popularity',ascending=True).head(10)
```

```python
sorted_df
```

```python
# perform all the statistics operation

df_track.describe().transpose()
```

```python
most_popular=df_track.query("popularity>90",inplace=False).sort_values('popularity',ascending=False)
```

```python
most_popular[:10]
```

```python
# cheak the artist name of index 

df_track[["artist_name"]].iloc[10]
```

```python
df_track["duration"]=df_track["duration_ms"].apply(lambda x:round(x/1000))
df_track.drop("duration_ms",inplace=True,axis=1)
```

```python
df_track.duration.head()
```

```python
corr_df = df_track.drop(["key", "mode"], axis=1).corr(method="pearson")
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df=df_track.sample(int(0.004*len(df_track)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="loudness",x="energy",color="c").set(title="loudness vs energy correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y="popularity",x="acousticness",color="y").set(title="popularity vs acousticness correlation")
```

```python
total_dr = df_track['duration']
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.barplot(x=df_track['popularity'], y=total_dr, ax=ax, errwidth=False)
ax.set_title("Popularity vs Duration")
plt.xticks(rotation=90)
plt.show()
```

```python
total_dr = df_track['duration']
fig_dims = (18, 7)
fig, ax = plt.subplots(figsize=fig_dims)
sns.lineplot(x=df_track['popularity'], y=total_dr, ax=ax)
ax.set_title("Popularity vs Duration")
plt.xticks(rotation=90)
plt.show()
```
